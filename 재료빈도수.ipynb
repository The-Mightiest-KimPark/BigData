## 학습재료 선정을 위한 레시피 재료 빈도

```python
import pandas as pd
# from konlpy.utils import pprint
import nltk
from konlpy.tag import Hannanum 
# import re


df = pd.read_csv("../output/코사인용2.csv")

df['ingredient_name'].fillna('', inplace=True)
data = df['ingredient_name'].tolist()
data = '\n'.join(data)

hannanum = Hannanum()
   
tokens_ko = hannanum.nouns(data)
ko_ = nltk.Text(tokens_ko)  
ko_.vocab()
freq_dic = dict(ko_.vocab())

sorted(freq_dic.items(), key=lambda x: x[1], reverse=True)
sortedDict = sorted(freq_dic.items(), key=lambda x: x[1], reverse=True)

print(sortedDict)
```
